```python
import sympy as sp

# 定义变量和函数
t, x, y = sp.symbols('t x y')
u = sp.Function('u')(t, x, y)
cx, cy = sp.symbols('cx cy')
dt, dx, dy = sp.symbols('Delta_t Delta_x Delta_y')
advection_eq = sp.Eq(u.diff(t), -cx * u.diff(x) - cy * u.diff(y))
# 定义离散点上的变量
u_ij_n = sp.Function('u')(t, x, y)
u_ij_n1 = sp.Function('u')(t + dt, x, y)
u_i_n_l = sp.Function('u')(t, x - dx, y)
u_j_n_l = sp.Function('u')(t, x, y - dy)
u_i_n_r = sp.Function('u')(t, x + dx, y)
u_j_n_r = sp.Function('u')(t, x, y + dy)
fin_x = (u_i_n_r - u_i_n_l) / (2*dx)
fin_y = (u_j_n_r - u_j_n_l) / (2*dy)
fin_t = (u_ij_n1 - u_ij_n) / dt
discretized_eq = advection_eq.subs({u.diff(x):fin_x, u.diff(y):fin_y,u.diff(t):fin_t})
sp.simplify(sp.solve(discretized_eq, u_ij_n1)[0])
```

```python
from examples.cfd import  plot_field
import numpy as np

def init_smooth(field, dx, dy):
    nx, ny = field.shape
    x = np.linspace(0, (nx-1)*dx, nx)
    y = np.linspace(0, (ny-1)*dy, ny)
    X, Y = np.meshgrid(x, y, indexing='ij')
    
    # 使用高斯分布初始化平滑场
    sigma = 0.2
    field[:] = np.exp(-((X-nx*dx/2)**2 + (Y-ny*dy/2)**2) / (2*sigma**2))
    
nt = 100  # Number of timesteps
dt = 0.01  # Timestep size (sigma=0.2)
c = 1  # Value for c

# Then we create a grid and our function
grid = Grid(shape=(3, 3), extent=(1., 1.))
u = TimeFunction(name='u', grid=grid,save = nt,time_order = 2, space_order = 2)

# We can now set the initial condition and plot it

#init_smooth(field=u.data[0], dx=grid.spacing[0], dy=grid.spacing[1])

u.data[0] = 0.1
u.data[0][1,1] = 1
plot_field(u.data[0])
```
